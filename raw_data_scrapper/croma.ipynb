```python
import tkinter as tk
import tkinter.filedialog as fd
from selenium.webdriver.common.by import By
# root = tk.Tk()
# file = fd.askopenfilename(parent=root, title='Choose a record file')
# file2 = fd.askopenfilenames(parent=root, title='Choose a pdf files')
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from IPython.display import clear_output
# root.destroy()
from time import sleep
import easygui
import time
import math
import numpy as np
import re
from datetime import date
from datetime import timedelta
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium import webdriver
import pandas as pd 
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import datetime
from dateutil.relativedelta import relativedelta

from selenium import webdriver
#set chromedriver.exe path
driver = webdriver.Chrome(executable_path="C:\chromedriver.exe")
driver.implicitly_wait(0)
#maximize browser
driver.maximize_window()
#launch URL
driver.get("https://www.croma.com/")
```

```python
driver.implicitly_wait(0)
# cc=['Frost Free refrigerator','Stationary Storage water heater','Colour TV','RAC (Variable speed)','TFL','LED Lamps','RAC (Fixed speed)','Direct cool refrigerator','Ceiling Fans']
cc=['RAC (Fixed speed)']

list=[
'https://www.croma.com/searchB?q=Air%20Conditioners%20Non%20Inverter%3Arelevance&text=Air%20Conditioners%20Non%20Inverter',
]

for w in range(len(list)):
    category=str(cc[w])
    driver.get(str(list[w]))
    
    #Get product link
    title=[]
    link=[]
    About=[]
    table=[]
    prodDescp=[]
    mainimg=[]
    while True:
        try:
            driver.find_element(by=By.CLASS_NAME, value='btn-viewmore').click()
        except:
            break
    time.sleep(1)
    job=driver.find_element(by=By.CLASS_NAME, value='content-wrap')
    soup=BeautifulSoup(job.get_attribute('innerHTML'),'html.parser')
    a=soup.find_all("a",class_="")
    for job_elem in a:
        try:
            link.append('https://www.croma.com'+job_elem['href'])
        except:
            pass
    c=soup.find_all("h3",class_="product-title plp-prod-title")
    for job_elem in c:
        title.append(job_elem.text)

    x = np.array(link)
    x = np.delete(x, np.arange(0, x.size, 3))
    x = np.delete(x, np.arange(0, x.size, 2))
    x=x.tolist()
    df = pd.DataFrame({
    'Title': title,
    'Link': x})

    #Get images link
    for j in range(0,len(df)):
        print(j)
        driver.get(df['Link'][j])
        time.sleep(1)
        
        try:
            driver.find_element(by=By.ID, value='custom-icon-close').click()
        except:
            pass
        
        try:
            img=[]
            job=driver.find_element(by=By.CLASS_NAME, value='swiper-wrapper')
            soup=BeautifulSoup(job.get_attribute('innerHTML'),'html.parser')
            a=soup.find_all("img",class_="pdp_lzy_img")
            for item in a:
                if(item['src'].split('.')[-1]!='gif'):
    #                 print(item['src'])
                    img.append(item['src'])
        except:
            img.append('none')
        mainimg.append(img)

        
        
        y=1000
        while True:
            try:
#                 print(y)
                driver.execute_script("window.scrollTo(0, "+str(y)+")")
                driver.find_element(by=By.ID, value='spec_viewMore_btn').click()
                break
            except:
                if(y>12000):
                    break
                else:
                    y=y+400
        y=2600        
        while True:
            try:
#                 print(y)
                driver.execute_script("window.scrollTo(0, "+str(y)+")")
                driver.find_element(by=By.ID, value='overview_viewMore_btn').click()
                break
            except:
                if(y>12000):
                    break
                else:
                    y=y+400
                
        try:
            #bullet points
            b=driver.find_element(by=By.CLASS_NAME, value='key-features-box').text
            About.append(b)
        except:
            About.append('none')
        try:
            #table
            b=driver.find_element(by=By.ID, value='specification_container').text
            table.append(b)
        except:
            table.append('none')
        try:
            #prodDescp
            b=driver.find_element(by=By.ID, value='overview_inner_container').text
            prodDescp.append(b)
        except:
            prodDescp.append('none')    

        
    df['bullet points']=0
    df['table']=0
    df['prodDescp']=0
    df['image link']=0
    df['bullet points']=About
    df['table']=table
    df['prodDescp']=prodDescp
    df['image link']=mainimg
    df.to_csv('final'+str(category)+'.csv')
#     break
```
